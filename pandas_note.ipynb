```python
tip = sns.load_dataset('tips')
tip.info()
```

```python
import statsmodels.api as sm
tip['sex'].replace(['Male','Female'],[0,1],inplace=True)
dfy0 = tip['tip']
dfx = tip['sex']
dfxx = sm.add_constant(dfx)
model = sm.OLS(dfy0, dfxx)
result = model.fit()
result.summary()
```
