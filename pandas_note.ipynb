```python
import pandas as pd
df = pd.read_csv('data/seoul_bicycle.csv')
df['대여일자'] = pd.to_datetime(df['대여일자'])
pd.to_numeric(df['운동량'], errors='coerce')
df['운동량']=pd.to_numeric(df['운동량'], errors='coerce')
bins = [0,200,400,df['운동량'].max()]
pd.cut(df['운동량'], bins = bins, labels=['운동부족','적당', '부족'], right=False)
df['운동량_qcut'] = pd.qcut(df['운동량'], q=10)
df
```

```python
import seaborn as sns
df = sns.load_dataset('titanic')
def trans_who(x):
    if x == 'man':
        return '남자'
    elif x == 'woman':
        return '여자'
    else:
        return '아이'
df['who'] = df['who'].apply(trans_who)
df['survived'] = df['survived'].apply(lambda x : '사망' if x==0 else '생존') 
df['survived'].value_counts()
```

```python
df['survived'] = df['survived'].apply(lambda x : 0 if x=='사망' else 1) 
df.groupby(['sex','pclass'])['survived'].mean()
```

```python
df.groupby(['sex','pclass'])['survived'].agg(['mean','sum']).reset_index()
```
